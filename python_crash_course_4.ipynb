```python
import matplotlib.pyplot as plt
from random import choice
class RandomWalk():
    """A class to generate to random walks"""
    def __init__(self,num_points=5000):
        """initialize attributes of a walk"""
        self.num_points = num_points
        # all walk start at (0,0)
        self.x_values = [0] # define two lists with initial value
        self.y_values = [0]
    def fill_walk(self):
        """Calculate all the points in the walk"""
        # keep taking steps untill the walk reaches the desired length
        while len(self.x_values)<self.num_points:
            # decide which direction to go and how far go to the direction
            x_direction = choice([1,-1]) # take a random value 1 or -1(pos or neg direction)
            x_distance = choice([0,1,2,3,4]) # take a random value from the list
            x_step = x_direction * x_distance
            
            y_direction = choice([1,-1])
            y_distance = choice([0,1,2,3,4]) # choice any value from the list
            y_step = y_direction * y_distance
            
            if x_step == 0 and y_step == 0: # if the steps are 0,then we don't need to add,the walk stops but we continue the loop to prevent this                
                continue
            next_x = self.x_values[-1] + x_step # we create next_step from last item of x_values list multiply with step
            next_y = self.y_values[-1] + y_step
            self.x_values.append(next_x) # append the next value in the 'x_values' list so that we can calculate 'next value'
            self.y_values.append(next_y)
while True:
    rw = RandomWalk(50000) # create a instance of a class store it to a variable
    rw.fill_walk()
    point_numbers = list(range(rw.num_points)) # range() to generate a list of numbers equal to the number of points in the walk,which we'll use to set the color of each poin in the walk
    
    plt.figure(dpi=128,figsize=(10,6))
    plt.scatter(rw.x_values,rw.y_values,c=point_numbers,cmap=plt.cm.Blues,edgecolor='none',s=1) # here 'c' means the higher the value of c,the deeper the color is,cmap is color map is blue,'s' means the density of each point
    plt.scatter(0,0,c='green',edgecolors='none',s=100) # here s means show the initial point largely,we plot point(0,0) in green in a larger size(s=100)
    plt.scatter(rw.x_values[-1],rw.y_values[-1],c='red',edgecolors='none',s=100) # we plot the last x and y value in the walk in red with a size of 100
    plt.show()
    keep_running = input("Make another walk? (y/n):")
    if keep_running == 'n':
        break
```

```python
from random import randint
class Die():
    """A class representing a single die"""
    def __init__(self,num_sides=6):
        """Assume a six-sided die"""
        self.num_sides = num_sides
    def roll(self):
        """Return a random value between 1 and number of sides"""
        return randint(1,self.num_sides)
die = Die()
results = []
for i in range(1000):
    result = die.roll()
    results.append(result)
#analyze the results
frequencies = []
for i in range(1,die.num_sides+1): # here num_sides of die is 6(define in class),and range is 1 to 7(within)
    frequency = results.count(i) # when i is 1/2 or 3, then count the all 'i' in the range store it to frequency variable
    frequencies.append(frequency)
print(frequencies)
```

```python
import csv
from datetime import datetime
from matplotlib import pyplot as plt
# Get dates and high temperatures from file.
filename = 'sitka_weather_2018_full.csv'
with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader) # next returns the head of csv file
    dates, highs, lows = [], [], []
    for row in reader:
        current_date = datetime.strptime(row[2],'%Y-%m-%d') # all row of 2nd column is date-time,and format is "y-m-d"
        dates.append(current_date)
        high = row[8] # all row of 8th column has temp_max column
        highs.append(high) # append 'max_temp' in 'highs' list
        low = row[9]
        lows.append(low)
# Plot data.
fig = plt.figure(dpi=128, figsize=(10, 6))
plt.plot(dates, highs, c='red',alpha=0.5)
plt.plot(dates,lows,c='blue',alpha=0.5)
plt.fill_between(dates,highs,lows,facecolor='blue',alpha=0.1)
# Format plot.
#plt.title("Daily high temperatures, July 2014", fontsize=24)
plt.xlabel('', fontsize=16)
fig.autofmt_xdate() # this one draws the date labels diagonally
plt.ylabel("Temperature (F)", fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=16)
plt.show()
```
