```python
from scipy import stats
import matplotlib.pyplot as plt
#generate some random variable form a non normal disb(log gamma disb which is pareto dis) and make QQ plot for this
fig = plt.figure()
ax1 = fig.add_subplot(211)
x = stats.loggamma.rvs(5,size=500) +5 #take loggamma disb(pareto distb)
prob = stats.probplot(x,dist=stats.norm,plot=ax1) #plotting a QQplot of pareto disb and normal dist
ax1.set_xlabel('')
ax1.set_title('Probplot against normal distribution')
plt.show()
```

```python
#emprical bootstrap based confidence interval
import numpy
from pandas import read_csv
from sklearn.utils import resample
from sklearn.metrics import accuracy_score
from matplotlib import pyplot
#load dataset
x = numpy.array([180,162,158,172,168,150,171,183,165,176])
#configure bootstrap
n_iterations = 1000 #total number of sampling
n_size = int(len(x)) # size of x,the resample size is less than or equal to x's size
#run bootstrap
medians = list() #initialize a list
for i in range(n_iterations): #upto all resamples
    s = resample(x,n_samples=n_size) #resample the data and the size is not more than original size
    m = numpy.median(s) #find a median of each resample data
    medians.append(m) #append the list
    
#plot score
pyplot.hist(medians)
pyplot.show()

#confidence intervals
alpha = 0.95
p = ((1.0-alpha)/2.0) * 100   #p is 2.5%
lower = numpy.percentile(medians,p) #find a median between of 2.5

p = (alpha+((1.0-alpha)/2.0)) *100  #p is 97.5%
upper = numpy.percentile(medians,p) #find a median greater of 97.5
print('%.lf confidence interval %.lf and %.lf' % (alpha*100,lower,upper))

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
d0 = pd.read_csv('./mnist_train.csv')
print(d0.head(5)) #print the first five rows of d0
#save the labels into a variable l
l = d0['label']
#drop the label feature and store the other coloumns in d
d = d0.drop("label",axis=1)
```

```python
labels = l.head(15000)
data = d.head(15000)
print("the shape of sample data = ",data.shape)
```

```python
#Data preprocessin: standardizing the data (remove mean from the data and divide by std dev and fianlly the value is between 0 and 1)
from sklearn.preprocessing import StandardScaler
standardized_data = StandardScaler().fit_transform(data)
print(standardized_data.shape)
```

```python
#find the co variance matrix whixh is: A^T *A, firstly standardized the data secondly make a co variance matrix
sample_data = standardized_data
#matrix multiplication using numpy
covar_matrix = np.matmul(sample_data.T,sample_data)
print("The shape of the co variance matrix = ",covar_matrix.shape)
```

```python
# finding the top two eigen-values and corresponding eigen-vectors for projecting 
# we just use the top two eigen values(782,783) for show the values in two dimension
#finding eigen value(lamda) and eigen vector(v)
from scipy.linalg import eigh
# the parameter 'eigvals' is defined (low value to high value)
# eigh function will return the eigen values in ascending order
# this code generates only top 2 (782 and 783) eigenvalues
values, vectors = eigh(covar_matrix,eigvals=(782,783))
print("Shape of the eigen vectors = ",vectors.shape)
# converting the eigen vectors into [2,d] shape for easyness of further computation
vectors = vectors.T
print("Updated shape of eigen vectors = ",vectors.shape)
```

```python
# here there are two dimension for 784 rows, so i have to do it for every rows(1500 rows) they have 2-D(two coloumns)
# projecting the original data sample on the plane
# formed by two principle eigen vectors by vector-vector multiplication
import matplotlib.pyplot as plt
new_coordinates = np.matmul(vectors,sample_data.T) # the dim of traspose sample data is (784,15000) and the dim of vector is (2,784)
print(new_coordinates.shape)
```

```python
# let's see these data which one converts 784 to 2_D 
import pandas as pd
#appending label to the 2nd projected data
new_coordinates = np.vstack((new_coordinates,labels)).T # vstack means vertical stack,here adding the new coordinates with label,there has 2 coordinates from 784 dim,now the dim of new_coordinates is (15000,2)
#creating a new data frame for plotting the labeled points.
dataframe = pd.DataFrame(data=new_coordinates,columns=("1st_principle","2nd_principle","labels"))
print(dataframe.head()) # head means just show first 5 values
```

```python
#plotting th 2D data points with seaborn
import seaborn as sn
sn.FacetGrid(dataframe,hue="labels",size=6).map(plt.scatter,'1st_principle','2nd_principle').add_legend() # show the dataframe using seaborn
plt.show()
# here show the two dim of 15k data which includes 0 to 9 labels
```

```python
#PCA using scikit-learn
#initializing the pca
# if we apply pca without using scikit learn function then we have to find out covariance matrix using matmul operation,before matmul
# operation we use standard scalar and then we use top two eigen values(782,283)
from sklearn import decomposition
pca = decomposition.PCA()
```

```python
# configuring the parameteres
# the number of components = 2
pca.n_components = 2 # take top two dimension
pca_data = pca.fit_transform(sample_data) # applying the PCA using scikit learn

# pca_reduced will contain the 2-d projects of simple data
print("shape of pca_reduced.shape = ", pca_data.shape)
```

```python
# attaching the label for each 2-d data point 
pca_data = np.vstack((pca_data.T, labels)).T  # apply the data in virtual stack,firstly transpose for label then again transpose extracting the original view shape

# creating a new data fram which help us in ploting the result data
pca_df = pd.DataFrame(data=pca_data, columns=("1st_principal", "2nd_principal", "label"))
sn.FacetGrid(pca_df, hue="label", size=6).map(plt.scatter, '1st_principal', '2nd_principal').add_legend() # using pca_df in seaborn for visualization 
plt.show()

```

```python
# PCA for dimensonality reduction, 
pca.n_components = 784
pca_data = pca.fit_transform(sample_data) # sample data is 15k standardized data
percentage_var_explained = pca.explained_variance_/ np.sum(pca.explained_variance_) #pca_explained_variance is lamda,find out the 'percentage_var_explained' by dividing all 'explained_variance'
cum_var_explained = np.cumsum(percentage_var_explained) #now sum all of the lamda's values,applying cumulutive sum for demonstrate purposes
#plot the PCA spectrum
plt.figure(1,figsize=(6,4))
plt.clf()
plt.plot(cum_var_explained,linewidth=2)
plt.axis('tight')
plt.grid()
plt.xlabel('n components')
plt.ylabel('Cumulative_explained_variance')
plt.show()
```

```python
# TSNE
from sklearn.manifold import TSNE


# picking the top 1000 points as TSNE takes a lot of time for 15k points
data_1000 = standardized_data[0:1000,:]  #take the data with standardization(means all values betwenn 0 and 1),using 'standardized_data' which is standardization already
labels_1000 = labels[0:1000] # take labels only
model = TSNE(n_components=2,random_state=0) #TSNE is randomized algorithm,it means always it re run the ans is slightly chnage for that reason you have to define the 'random_state',n_components means the dim which one we use
#the number of components =2
# default perplexity = 30
# default learning rate =200
# default maximum number of iterations for the optimization = 1000
tsne_data = model.fit_transform(data_1000) # the whole 1000 data and 786 dim converts to 2 dim,because we define the n_components is 2

# creating a new data frame which help us in plotting the result data
tsne_data = np.vstack((tsne_data.T,labels_1000)).T # here creating a vertical stack where stands data alongside with labels and make it tranpose(vstack means vertical stack) and again transpose to convert it original format
tsne_df = pd.DataFrame(data=tsne_data,columns=("Dim_1", "Dim_2", "label")) # here creating a dataframe on tsne_data

#plotting the result with tsne
sn.FacetGrid(tsne_df,hue="label",size=6).map(plt.scatter,'Dim_1','Dim_2').add_legend()
plt.show()
```

```python
#now change perplexity to 50 and iteration is 5000, perplexity means the no of neighbouring point takes each point
model = TSNE(n_components=2, random_state=0,perplexity=50,n_iter=5000)
tsne_data = model.fit_transform(data_1000)
#creating a new data frame which help us in plotting the result data
tsne_data = np.vstack((tsne_data.T,labels_1000)).T
tsne_df = pd.DataFrame(data=tsne_data,columns=("Dim_1", "Dim_2","label"))
#plotting the result with tsne
sn.FacetGrid(tsne_df,hue="label",size=6).map(plt.scatter,'Dim_1','Dim_2').add_legend()
plt.title('With perplexity = 50 , n_iter=5000')
plt.show()
```

```python
#now change perplexity to 50
model = TSNE(n_components=2, random_state=0,perplexity=2,n_iter=1000) # here perplexity is very low,each point takes just neighbouring 2 points,so it is underfitting
tsne_data = model.fit_transform(data_1000)
#creating a new data frame which help us in plotting the result data
tsne_data = np.vstack((tsne_data.T,labels_1000)).T
tsne_df = pd.DataFrame(data=tsne_data,columns=("Dim_1", "Dim_2","label"))
#plotting the result with tsne
sn.FacetGrid(tsne_df,hue="label",size=6).map(plt.scatter,'Dim_1','Dim_2').add_legend()
plt.title('With perplexity = 2 , n_iter=1000')
plt.show()
```
